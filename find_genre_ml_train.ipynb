```python
import nltk
# Do this the first time implmenting in production machine
# nltk.download()
from nltk.tag import pos_tag, map_tag
from nltk.corpus import stopwords

import pandas as pd
import ast
import wikipedia
import re
import wiki_table_extract as wt
```

```python
def plot_to_words( raw_review ):
    # Function to convert a raw review to a string of words
    # The input is a single string (a raw movie review), and 
    # the output is a single string (a preprocessed movie review)
    #
    # 1. Remove HTML
    #review_text = BeautifulSoup(raw_review).get_text() 
    #
    # 2. Remove non-letters        
    letters_only = re.sub("[^a-zA-Z]", " ", raw_review) 
    #
    # 3. Convert to lower case, split into individual words
    words = letters_only.lower().split()                             
    #
    # 4. In Python, searching a set is much faster than searching
    #   a list, so convert the stop words to a set
    stops = set(stopwords.words("english"))                  
    # 
    # 5. Remove stop words
    meaningful_words = [w for w in words if not w in stops]   
    #
    # 6. Join the words back into one string separated by space, 
    # and return the result.
    return( " ".join( meaningful_words ))
```

```python
def plot_to_nlp( plot_text ):
    
    #bigrams = [b for l in plot_text for b in zip(l.split(" ")[:-1], l.split(" ")[1:])]
    
    text = nltk.word_tokenize(plot_text)
    posTagged = pos_tag(text)
    simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]

    w_nlp = ""
    for wcombo in simplifiedTags:
        w_nlp = w_nlp + " " + wcombo[1]
    
    return w_nlp
```

```python
# Find Genre for movies that are missing Ganre information.
# Option 1. Get the Wiki Link. Search Summary and get Genre.
# Option 2. Use Ml to predict if Plot info available.

movieListDF = pd.read_csv("all_tamil_movies_1.csv")
movieListDF =  movieListDF[['Genre','Title']]
movieListDF = movieListDF.dropna(subset=['Genre'])

#print (movieListDF.head())

allGenrePlotColumns = ['Genre','OriginalPlot']
dfPlotGenreList = pd.DataFrame(columns=allGenrePlotColumns)

for index,row in movieListDF.iterrows():
    try:
        
        titleWikiDict = ast.literal_eval(row['Title'])
        titleWikiLink = (next (iter (titleWikiDict.values())))
        titleWiki = (next (iter (titleWikiDict.keys())))
        genreWikiDict = ast.literal_eval(row['Genre'])
        genreWikiLink = (next (iter (genreWikiDict.keys()))).lower()

        if titleWikiLink.startswith('/wiki/'):

            # Print Title Wiki Link
            print (titleWikiLink,"-",genreWikiLink)
            wikiCompLink = titleWikiLink.replace('/wiki/','')
            wikipage=wikipedia.WikipediaPage(wikiCompLink)
            wikiPlot = wikipage.section('Plot')
            wikiSummary = wikipage.summary
            wikiCast = wikipage.section('Cast')
            
            if wikiPlot is not None:
            
                row_list = []
                wikidict = {}
                wikidict['MoviePage'] = titleWikiLink            
                wikidict['Genre'] = genreWikiLink
                wikidict['OriginalPlot'] = wikiPlot
                wikidict['Summary'] = wikiSummary
                wikidict['Cast'] = wikiCast

                # Will be filled later
                wikidict['BOW_Plot'] = ""
                wikidict['NLP_Plot'] = ""
                wikidict['Combined_Plot'] = ""

                # Rowlist to be appended
                row_list.append(wikidict)
                dfPlotGenreList = dfPlotGenreList.append(row_list)
        
    except Exception:
        print ("Exception Caught",Exception)
        pass

for index,row in dfPlotGenreList.iterrows():
        wikiPlot = row['OriginalPlot']
        row['BOW_Plot'] = plot_to_words (wikiPlot)    
        row['NLP_Plot'] = plot_to_nlp (wikiPlot)
        row['Combined_Plot'] = row['BOW_Plot'] + row['NLP_Plot']
```

```python
dfPlotGenreList.to_csv('genre_plot.csv',sep='\t')
```

```python
from sklearn.model_selection import train_test_split

X_train_all, X_test_all, y_train, y_test = train_test_split(dfPlotGenreList[['MoviePage','Combined_Plot']], dfPlotGenreList['Genre'], test_size=0.33, random_state=42)
X_train = X_train_all['Combined_Plot']
X_test = X_test_all['Combined_Plot']
```

```python
clean_train_plots = []
for row in X_train:
    clean_train_plots.append(row)
```

```python
print ("Creating the bag of words")
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the "CountVectorizer" object, which is scikit-learn's
# bag of words tool.  
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000) 

# fit_transform() does two functions: First, it fits the model
# and learns the vocabulary; second, it transforms our training data
# into feature vectors. The input to fit_transform should be a list of 
# strings.
train_data_features = vectorizer.fit_transform(clean_train_plots)

# Numpy arrays are easy to work with, so convert the result to an 
# array
train_data_features = train_data_features.toarray()
```

```python
import numpy as np

# Take a look at the words in the vocabulary
vocab = vectorizer.get_feature_names()
print (vocab)

# Sum up the counts of each vocabulary word
dist = np.sum(train_data_features, axis=0)

# For each, print the vocabulary word and the number of times it 
# appears in the training set
for tag, count in zip(vocab, dist):
    print(count,tag)
```

```python


print ("Training the random forest")
from sklearn.ensemble import RandomForestClassifier

# Initialize a Random Forest classifier with 100 trees
forest = RandomForestClassifier(n_estimators = 200) 

# Fit the forest to the training set, using the bag of words as 
# features and the sentiment labels as the response variable
#
# This may take a few minutes to run
forest = forest.fit( train_data_features, y_train )

```

```python

# Create an empty list and append the clean reviews one by one

clean_test_plots = []
for row in X_test:
    clean_test_plots.append(row)

# Get a bag of words for the test set, and convert to a numpy array
test_data_features = vectorizer.transform(clean_test_plots)
test_data_features = test_data_features.toarray()

# Use the random forest to make sentiment label predictions
result = forest.predict(test_data_features)

# Copy the results to a pandas dataframe with an "id" column and
# a "sentiment" column
output = pd.DataFrame( data={"movie_page":X_test_all['MoviePage'],"Genre":y_test, "predicted":result} )

# Use pandas to write the comma-separated output file
output.to_csv( "Bag_of_Words_model.csv", index=False)

```
